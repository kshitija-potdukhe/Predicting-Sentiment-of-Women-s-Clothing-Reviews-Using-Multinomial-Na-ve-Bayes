### Title of Project
**Predicting Sentiment of Women’s Clothing Reviews Using Multinomial Naïve Bayes**

### Objective
To classify customer reviews of women’s clothing into different sentiment categories (e.g., positive, neutral, negative) using Multinomial Naïve Bayes for predictive analysis.

### Data Source
You can use datasets from sources like Kaggle, UCI Machine Learning Repository, or any available review dataset with labeled sentiments. For instance, Kaggle’s "Clothing Reviews Dataset."

### Import Library
```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

### Import Data
```python
# Replace 'file_path' with your actual file path
data = pd.read_csv('file_path.csv')
```

### Describe Data
```python
print(data.head())
print(data.info())
print(data.describe())
print(data['sentiment'].value_counts())
```

### Data Visualization
```python
sns.countplot(x='sentiment', data=data)
plt.title('Sentiment Distribution')
plt.show()
```

### Data Preprocessing
```python
# Remove missing values
data.dropna(subset=['review', 'sentiment'], inplace=True)

# Text cleaning (optional, depending on your data)
data['review'] = data['review'].str.lower().str.replace('[^\w\s]', '')

# Tokenization and vectorization
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(data['review'])
```

### Define Target Variable (y) and Feature Variables (X)
```python
X = vectorizer.transform(data['review'])
y = data['sentiment']
```

### Train Test Split
```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### Modeling
```python
model = MultinomialNB()
model.fit(X_train, y_train)
```

### Model Evaluation
```python
y_pred = model.predict(X_test)
print('Accuracy:', accuracy_score(y_test, y_pred))
print('Classification Report:\n', classification_report(y_test, y_pred))
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred))
```

### Prediction
```python
# Example of predicting a new review
new_review = ["This dress is amazing and fits perfectly!"]
new_review_transformed = vectorizer.transform(new_review)
prediction = model.predict(new_review_transformed)
print('Predicted Sentiment:', prediction[0])
```

### Explanation
The Multinomial Naïve Bayes model is used for text classification based on the probability of words in each sentiment class. We preprocess the text data by tokenizing and vectorizing, then split the data for training and testing. The model is evaluated using metrics such as accuracy and the confusion matrix to assess its performance. Predictions can then be made on new reviews to classify their sentiment.
